```python
import matplotlib.pyplot as plt
from NeuralNetwork import *
from sklearn.datasets import make_circles
```

```python
# Data train | test 

X_train, Y_train = make_circles(n_samples=1000, noise=0.1, factor=0.3, random_state=0)
X_test, Y_test = make_circles(n_samples=500, noise=0.1, factor=0.3, random_state=5)
X_train, X_test = X_train / X_train.max(), X_test / X_test.max()
Y_train, Y_test = Y_train.reshape(Y_train.shape[0], 1), Y_test.reshape(Y_test.shape[0], 1)
print("Training DataSet Input:", X_train.shape)
print("Training DataSet Output:", Y_train.shape)
print("Testing DataSet Input:", X_test.shape)
print("Testing DataSet Output:", Y_test.shape)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.title("2D Train Input")
plt.scatter(X_train[:, 0], X_train[:, 1], c=Y_train, cmap='summer')
plt.subplot(1,2,2)
plt.title("2D Test Input")
plt.scatter(X_test[:, 0], X_test[:, 1], c=Y_test, cmap='summer')
plt.show()
```

```python
# Basic model 

neuralNetwork = NeuralNetwork(nInput=2, nOutput=1, hiddenNxM=[2, 2], rand_seed=0)
result = neuralNetwork.calibrate(X_train.T, Y_train.T, X_test.T, Y_test.T, learning_rate=0.01, iteration=10000, n_sync=10)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(2,2,1)
plt.title("Train")
plt.plot(result['train_Loss'])
plt.subplot(2,2,2)
plt.title("Test")
plt.plot(result['test_Loss'])
plt.subplot(2,2,3)
plt.plot(result['train_Acc'])
plt.subplot(2,2,4)
plt.plot(result['test_Acc'])
plt.show()
```
